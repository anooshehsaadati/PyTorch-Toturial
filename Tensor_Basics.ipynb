## **Import important Libraries**

```python
import torch
import numpy as np
```

### **create Tensor in Torch**

```python
# Empty tensor with 3 dimensions
x1 = torch.empty(4, 2, 3)
print(x1)
```

```python
# Random tensor with 2 dimensions
x2 = torch.rand(2, 2)
print(x2)
```

```python
# Zeroes tensor with 2 dimensions
x3 = torch.zeros(2, 2)
print(x3)
```

```python
# Ones tensor with 2 dimensions
x4 = torch.ones(2, 2)
print(x4)
```

### **Type of Tensor**

```python
# default type of tensor is float32
x1.dtype
```

```python
# you can set other type to tensor
x5 = torch.ones(2, 2, dtype=torch.int) # torch.int, torch.double, torch.float16
print(x5.dtype)
```

### **Size of Tensor**

```python
x6 = torch.rand(2, 2)
print(x6.size())
```

### **Create tensor with list of data**

```python
x7 = torch.tensor([2.5, 3, 1])
print(x7)
```

### **Basic Operations**

```python
x8 = torch.rand(2, 2)
x9 = torch.rand(2, 2)

print(x8, x9)

# Addition
x10 = x9 + x8
x10 = torch.add(x9, x8)
x9.add_(x8)
print(x10, x9)

# Subtraction
x11 = x9 - x8
x11 = torch.sub(x9, x8)
x9.sub_(x8)
print(x11, x9)

# Multiplication
x12 = x9 * x8
x12 = torch.mul(x9, x8)
x9.mul_(x8)
print(x12, x9)

# Division
x12 = x9 / x8
x12 = torch.div(x9, x8)
x9.div_(x8)
print(x12, x9)
```

### **Tensor Slicing**

```python
x13 = torch.rand(5, 3)
print(x13)

print(x13[:, 0]) # All rows and column 0

print(x13[1, :]) # row 1 and all columns

print(x13[1, 1]) # actual element
print(x13[1, 1].item()) # if tensor have one element, you can use .item() to get actual value
```

### **Tensor Reshaping**

```python
x14 = torch.rand(4, 4)
print(x14)

x15 = x14.view(16)
print(x15)

x16 = x14.view(-1, 8) # -1 is automatically to right dimension -- if multiply of dimension not divisible by input, get error like x14.view(-1, 3)
print(x16, x16.size())
```
